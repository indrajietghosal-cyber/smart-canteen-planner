```python
import numpy as np
import pandas as pd

np.random.seed(42)

days = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
time_slots = ["Breakfast","Lunch","Snacks","Dinner"]
items = ["Idli", "Dosa", "Samosa", "Chai", "Coffee", "Fried Rice"]

data = []

for day in days:
    for time in time_slots:
        for item in items:

            # Basic demand logic
            base = np.random.randint(20, 60)
            if time == "Lunch":
                base += 40
            if item in ["Chai", "Coffee"] and time == "Snacks":
                base += 30

            demand = base + np.random.randint(-10, 10)

            row = {
                "day_of_week": day,
                "time_slot": time,
                "item": item,
                "historical_orders": base,
                "item_popularity": np.random.uniform(0.3, 0.9),
                "is_exam_week": np.random.choice([0,1], p=[0.8,0.2]),
                "special_event_flag": np.random.choice([0,1], p=[0.9,0.1]),
                "weather": np.random.choice(["Normal","Rainy"], p=[0.7,0.3]),
                "queue_length": np.random.randint(5, 40),
                "active_counters": np.random.randint(1, 4),
                "staff_count": np.random.randint(3, 10),
                "avg_prep_time": np.random.uniform(3, 12),
                "profit_margin": np.random.uniform(0.2, 0.9),
                "ingredient_availability": np.random.uniform(0.5, 1.0),
                "prep_time": np.random.uniform(5, 15),
                "wastage_history": np.random.uniform(0.05, 0.3),
                "demand_qty": demand,                 # Model 1 target
                "expected_wait_time": demand / 10,    # Model 2 target
                "prep_level": np.random.choice(["High","Medium","Low"])
            }

            data.append(row)

df = pd.DataFrame(data)
df.head()
```

```python
df.to_csv("../data/canteen_data.csv", index=False)
```
